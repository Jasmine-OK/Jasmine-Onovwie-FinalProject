```python
import pandas as pd

# Define the column names manually
columns = ['age', 'gender', 'income', 'utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']

# Load the CSV with the defined column names
df = pd.read_csv('user_data.csv', header=None, names=columns)

print(df.head())  # Check if the data is correctly loaded now
```

```python
import matplotlib.pyplot as plt

# Show ages with the highest income
df.groupby('age')['income'].sum().plot(kind='bar', color='pink')
plt.title('Total Income by Age')
plt.xlabel('Age')
plt.ylabel('Total Income')
plt.show()
```

```python
# Sum of expenses for each gender across all spending categories
df.groupby('gender').sum()[['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']].plot(kind='bar', stacked=True)

plt.title('Spending by Gender Across Categories')
plt.xlabel('Gender')
plt.ylabel('Total Spending')
plt.legend(title='Spending Categories')
plt.show()
```

```python
# Show ages with the highest income
df.groupby('age')['income'].sum().plot(kind='bar', color='pink')
plt.title('Total Income by Age')
plt.xlabel('Age')
plt.ylabel('Total Income')

# Save the chart as a PNG image
plt.savefig('income_by_age.png')

plt.show()
```

```python
# Sum of expenses for each gender across all spending categories
df.groupby('gender').sum()[['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']].plot(kind='bar', stacked=True)

plt.title('Spending by Gender Across Categories')
plt.xlabel('Gender')
plt.ylabel('Total Spending')
plt.legend(title='Spending Categories')

# Save the chart as a PNG image
plt.savefig('spending_by_gender.png')

plt.show()
```
